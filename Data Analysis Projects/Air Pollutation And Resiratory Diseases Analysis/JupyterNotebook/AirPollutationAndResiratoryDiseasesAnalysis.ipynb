**For air poplation, I found six datasets from data.gov.sg which shows the mean content in microgram per cubic meter of the pollutants.(six types of air pollutants are carbon monoxide, nitrogen dioxide, pm 10(inhalable particles that are 10 micron diameter or less), pm 2.5(inhalable particles that are 2.5 micron diameter or less),ozone and sulphur dioxide)**

```python
import pandas as pd
import matplotlib.pyplot as plt
# writing a function to describe the number of columns and rows,name of the columns, number of duplicate rows and index,
#the data types of each column,
#the number of null values in each column and the top three rows  of each data set. 


def describeDataSet(dframe):
    print(f'Number of rows: {dframe.shape[0]}\nNumber of columns: {dframe.shape[1]}\n')
    print(f'Column Names are:')
    names = list(dframe.columns);
    for num,name in enumerate(names):
        print(f'{num+1}. {name}')
    df_duplicated = dframe[dframe.duplicated()]
    if(df_duplicated.shape[0]==0):
        print('\nThis data set has no duplicate rows.\n')
    else:
        print(f'\nThe index(s) of duplicate rows {list(df_duplicated.index)}\n')
    print(f'The data type of this data set are \n{dframe.dtypes}\n')
    print(f'The number of null values\n{dframe.isnull().sum()}\n')
    print(f'The top three rows of data set are:\n{dframe.head(3)}')


```

**Now I will set year as indexes for the air pollutation data frames and then I will merge these data sets as one data frame and rename the columns(to set the column name as label in line graph) and the purpose of doing these is to create a line graph to get a better understanding of the trend(highs and lows) of these air pollutants content in Singapore.** 

```python
#setting the year column as index for the above data frames 
# wrting a function which makes the first column of each dataframe as index
def SetIndex(*datasets):
    new_datasets = []
    for dataset in datasets:
        dataset = dataset.set_index(dataset.columns[0])
        new_datasets.append(dataset)
    return tuple(new_datasets)
df_CO,df_NO2,df_Ozone,df_pm10,df_pm2_5,df_SO2 = SetIndex(df_CO,df_NO2,df_Ozone,df_pm10,df_pm2_5,df_SO2)
```

```python
# merging these data frame into one dataframe
#function for merging data sets on their indexes in this case, year
def MergeDataSetsOnIndex(*datasets):
    merged_dataset = datasets[0]
    for dataset in datasets[1:]:
        merged_dataset = pd.merge(merged_dataset,dataset,how='outer',left_index=True,right_index=True)
    return merged_dataset
df_airpollutants = MergeDataSetsOnIndex(df_CO,df_NO2,df_Ozone,df_pm10,df_pm2_5,df_SO2)
```

## Finding The relationship between the air pollutants and polyclinic attendance for Acute Upper Respiratory Tract infections

**For this task, I will create a data frame with the values of air pollutants and the number of attendance by merging two data frame df_actue_resipratory and df_airpollutants.**

```python
# creating the data frame to draw a scatter plot and to calculate the correlations
#merging two data frames df_airpollutants and df_acute_respiratory
df_airpollutants.reset_index(inplace = True)
df_airpollutants['year'] = df_airpollutants['year'].astype(int)
df_actue_resipratory['year'] = df_actue_resipratory['year'].astype(int)
df_for_correlation = pd.merge(df_actue_resipratory, df_airpollutants, how='inner', on='year')
```

```python
df_for_correlation
```

```python
#merging with another data set, df_respiratory_hospitalization
df_for_correlation = pd.merge(df_for_correlation, df_respiratory_hospitalization, how='inner', on='year')
df_for_correlation
```

```python
df_for_correlation.drop(columns=['disease','disease_condition'],inplace=True)
df_for_correlation
```

### Creating a scatter plot between the air pollutants and number of cases with acute respiratory tract infections

**Now, I will create two scatter plots, one for between CO,NO2,PM2.5 and number of cases and the other one for between Ozone, PM10, SO2 and number of cases. This is because the CO,NO2 and PM 2.5 have small number of microgram per cubic meter and the other three have high number so to see clear relationship I create two scatter plot.**

```python
# creating a function for a scatter plot 
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 6))
def scatterPlot(ax,data):
    for column in data:
        ax.scatter(data[column], df_for_correlation['no._of_cases'],label=column)
    ax.set_xlabel('Microgram per cubic meter (Air pollutant content)')
    ax.set_ylabel('Number of Acute Upper Respiratory Infection Cases')
    ax.legend()
#splitting columns
data1 = df_for_correlation.iloc[:,[2,3,6]]
data2 = df_for_correlation.iloc[:,[4,5,7]]
scatterPlot(axes[0],data1)
axes[0].set_title('Scatter Plot for Air Pollutant(CO,NO2,PM2.5,) vs Number of Actue Respiratory Disease Cases')
scatterPlot(axes[1],data2)
axes[1].set_title('Scatter Plot for Air Pollutant(Ozone,PM10,SO2) vs Number of Actue Respiratory Disease Cases')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

**From the analysis of the above scatter plots,NO2 seems to have a strong positive relationship and for other data points the relationship is not clear, so I will try to find the perason correlation coefficient**

```python
import numpy as np
columns = ['CO', 'NO2', 'Ozone', 'PM10', 'PM2.5', 'SO2']


# Calculate correlation matrices
correlation_cases = df_for_correlation[columns].corrwith(df_for_correlation['no._of_cases'])
correlation_percentage = df_for_correlation[columns].corrwith(df_for_correlation['percentage_discharges'])

print(f'The correlation between air pollutants and number of cases of actue respiratory disease :\n{correlation_cases}\n')
print(f'The correlation between air pollutants and percentage of hospitalization with respiratory disease :\n{correlation_percentage}\n')
```

**From correlation between air pollutants and number of cases of actue respiratory disease, there some positive relationships but for the correlation between air pollutants and percentage of hospitalization with respiratory disease, most of them have negative relationships, so this does not align with the assumption that the number of cases might increase with the increase in air pollutant. So I will omit the correlation between air pollutants and percentage of hospitalization with respiratory disease.** 

## Creating a heat map for The correlation between air pollutants and number of cases of actue respiratory disease

```python
# The following heat map is referenced from https://python-charts.com/correlation/heatmap-matplotlib/
import numpy as np
correlation_df = pd.DataFrame(correlation_cases).T


fig, ax = plt.subplots(figsize=(8, 2))
cax = ax.imshow(correlation_df, cmap='coolwarm', vmin=-1, vmax=1, aspect='auto')


ax.set_xticks(np.arange(len(correlation_df.columns)))
ax.set_yticks([0])
ax.set_xticklabels(correlation_df.columns)
ax.set_yticklabels(['No. of Cases'])

#printing number inside the color cells
for i in range(len(correlation_df.columns)):
    text = ax.text(i, 0, f'{correlation_df.iloc[0, i]:.2f}', ha='center', va='center', color='black')

#creating the colorbar 
cbar = fig.colorbar(cax, orientation='vertical')

# Add labels and title
plt.xlabel('Air Pollutants')
plt.title('Pearson Correlation Heatmap between Airpollutants and Number of cases for Acute Respiratory Disease Cases')

plt.show()
```

**From the above heat map, NO2 is has strong positive correlationship, and SO2 has moderate positive correlationship and PM2.5 also have weak positive  coorelatinship and CO,Ozone and PM10 has very weak correlationship.**
